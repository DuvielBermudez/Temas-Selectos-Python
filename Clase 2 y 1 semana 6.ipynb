```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruta="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/CEMEXCPO.csv"
```

```python
df = pd.read_csv(ruta)
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date']) #convertir el dato a una serie de tiempo
df.info()
```

para que cambie el índice que pone por default y lo sustituya por la columna que indico

```python
df = df.set_index("Date")
df
```

```python
df.loc[df.idxmax()[3]][3]
```

```python
df.loc[df.idxmin()[3]][3]
```

Para graficar podemos usar el siguiente codigo

```python
plt.figure(figsize=(20,10), dpi=150)
plt.plot(df.index,df["Close"],lw=2, color='black')
plt.xlabel('Fecha', fontdict = {'fontsize':25 })
plt.ylabel('Precio ($)',fontdict = {'fontsize':25 })
plt.title('Precio de Cierre (CEMEX)', fontsize = 25)
plt.tick_params(labelsize=20)
plt.grid(True)

print('Precio de cierre más bajo: ',df.loc[df.idxmin()[3]][3], ', fecha: ',df.idxmin()[3])
print('Precio de cierre más alto: ',df.loc[df.idxmax()[3]][3], ', fecha: ',df.idxmax()[3])
```

Para un estilo mas profesional se le puede pedir ayuda a una inteligencia artificial y graficacar

```python
# Estilo general más profesional
plt.style.use('seaborn-v0_8-whitegrid')  # mejora la estética del gráfico

plt.figure(figsize=(16, 8), dpi=150)

# ==========================
# LÍNEAS PRINCIPALES
# ==========================
plt.plot(df.index, df["Close"], lw=2.5, color="black", label="Precio de cierre")  # línea principal


# ==========================
# PUNTOS CLAVE (MÍNIMO y MÁXIMO)
# ==========================
min_idx = df["Close"].idxmin()  # índice del valor mínimo
max_idx = df["Close"].idxmax()  # índice del valor máximo
min_val = df.loc[min_idx, "Close"]
max_val = df.loc[max_idx, "Close"]

# Marcamos los puntos en el gráfico
plt.scatter(min_idx, min_val, color='red', s=80, label=f'Mínimo: ${min_val:.2f}')
plt.scatter(max_idx, max_val, color='green', s=80, label=f'Máximo: ${max_val:.2f}')

# Anotaciones con valores
plt.annotate(f'Mínimo\n${min_val:.2f}',
             xy=(min_idx, min_val),
             xytext=(-100, -40),
             textcoords='offset points',
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=14, color='red')

plt.annotate(f'Máximo\n${max_val:.2f}',
             xy=(max_idx, max_val),
             xytext=(40, 40),
             textcoords='offset points',
             arrowprops=dict(facecolor='green', arrowstyle='->'),
             fontsize=14, color='green')

# ==========================
# CONFIGURACIÓN DE ETIQUETAS Y TÍTULOS
# ==========================
plt.xlabel('Fecha', fontsize=18, labelpad=10)
plt.ylabel('Precio ($)', fontsize=18, labelpad=10)
plt.title('Precio de Cierre - CEMEX', fontsize=22, fontweight='bold', pad=15)


# ==========================
# FORMATOS DE EJES Y LEYENDAS
# ==========================
plt.grid(alpha=0.3)
plt.tick_params(labelsize=14)
plt.legend(fontsize=14)
plt.tight_layout()

plt.show()

# Mostrar también los datos por consola
print(f"📉 Precio de cierre más bajo: ${min_val:.2f}, fecha: {min_idx}")
print(f"📈 Precio de cierre más alto: ${max_val:.2f}, fecha: {max_idx}")
```

TAREA DE LA PRIMERA SESION SEMANA 6 REALIZAR EL EJERCICIO DEL WALMART QUE SE ENCUENTRA EN EL GITHUB

```python
walmart="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/WALMEX.csv"
wa = pd.read_csv(walmart) #lee el archivo como un csv
wa.head()
```

```python
wa["Date"] = pd.to_datetime(wa["Date"])
wa = wa.set_index("Date")
wa.info()
```

```python
wa
```

```python
# Estilo general más profesional
plt.style.use('seaborn-v0_8-whitegrid')  # mejora la estética del gráfico

plt.figure(figsize=(16, 8), dpi=150)

# ==========================
# LÍNEAS PRINCIPALES
# ==========================
plt.plot(wa.index, wa["Close"], lw=2.5, color="black", label="Precio de cierre")  # línea principal

# Promedio móvil (ventana de 20 días) para ver la tendencia
wa["MA20"] = wa["Close"].rolling(window=20).mean()
plt.plot(wa.index, wa["MA20"], color="blue", lw=2, linestyle="--", label="Media móvil (20 días)")

# ==========================
# PUNTOS CLAVE (MÍNIMO y MÁXIMO)
# ==========================
min_idx = wa["Close"].idxmin()  # índice del valor mínimo
max_idx = wa["Close"].idxmax()  # índice del valor máximo

# Marcamos los puntos en el gráfico
plt.scatter(min_idx, wa["Close"].min(), color="red", s=80, zorder=5, label="Valor mínimo")
plt.scatter(max_idx, wa["Close"].max(), color="green", s=80, zorder=5, label="Valor máximo")

# Anotaciones con valores
plt.annotate(f"Mín: ${wa['Close'].min():.2f}",
             xy=(min_idx, wa["Close"].min()),
             xytext=(-80, -30), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="red", lw=1.5),
             fontsize=12, color="red", fontweight="bold")

plt.annotate(f"Máx: ${wa['Close'].max():.2f}",
             xy=(max_idx, wa["Close"].max()),
             xytext=(-80, 30),  # misma distancia pero hacia arriba
             textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="green", lw=1.5),
             fontsize=12, color="green", fontweight="bold")

# ==========================
# SOMBREADO ENTRE PRECIO ALTO Y BAJO (SI EXISTEN)
# ==========================
if "High" in wa.columns and "Low" in wa.columns:
    plt.fill_between(wa.index, wa["Low"], wa["High"], color="gray", alpha=0.15, label="Rango diario")

# ==========================
# CONFIGURACIÓN DE ETIQUETAS Y TÍTULOS
# ==========================
plt.xlabel("Fecha", fontsize=18, labelpad=10)
plt.ylabel("Precio ($)", fontsize=18, labelpad=10)
plt.title("Evolución del precio de cierre - WALMART", fontsize=22, fontweight="bold", pad=15)

# ==========================
# FORMATOS DE EJES Y LEYENDAS
# ==========================
plt.grid(alpha=0.3)
plt.tick_params(labelsize=14)
plt.legend(fontsize=14)
plt.tight_layout()


# Añadimos línea horizontal en el promedio general
plt.axhline(wa["Close"].mean(), color="orange", lw=1.5, linestyle=":", label="Promedio general")

plt.tight_layout()  # evita solapamiento de etiquetas
plt.show()

```

```python
ruta="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/contam2022.csv"
```

```python
contam = pd.read_csv(ruta, parse_dates = True, index_col=0)
contam.head()
```

```python
contam.info()
```

```python
contam.columns = ['ozono','dioxido_azufre',
                  'dioxido_nitrogeno','monoxido_carbono',
                  'PM10']
contam.head()
```

```python
contam.plot(subplots=True)
plt.show()
```

```python
np.sqrt(len(contam))
```

```python
plt.figure(figsize=(5,4))
plt.hist(contam['dioxido_nitrogeno'], bins=19,
         color='green', edgecolor='black')
plt.title('Dioxido de nitrogeno')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(2,3,2)

plt.hist(contam['ozono'], color='skyblue', edgecolor='black')
plt.title('O3')
plt.ylabel('Frecuencia')

plt.subplot(2,3,1)
plt.hist(contam['dioxido_azufre'], color='skyblue', edgecolor='black')
plt.title('SO2')
plt.ylabel('Frecuencia')

plt.subplot(2,3,3)
plt.hist(contam['dioxido_nitrogeno'], color='skyblue', edgecolor='black')
plt.title('NO2')
plt.ylabel('Frecuencia')

plt.subplot(2,3,4)
plt.hist(contam['monoxido_carbono'], color='skyblue', edgecolor='black')
plt.title('CO')
plt.ylabel('Frecuencia')

plt.subplot(2,3,5)
plt.hist(contam['PM10'], color='skyblue', edgecolor='black')
plt.title('PM10')
plt.ylabel('Frecuencia')

plt.tight_layout()# para que no se solapen las graficas y se puedan orientar equitativamente
plt.show() # para mostrar las graficas
```

Para evaluar si un conjunto de datos tiene destribucion normal se usa la prueba de shapiro-wilk esta es una prueba de hipotesis

```python
#se llama una libreria que nos ayuda a hacer analisis estadistico mas avanzados
from scipy.stats import shapiro
for col in contam.columns:
    stat, p = shapiro(contam[col])
    print(f"{col}: Estadístico = {stat:.3f}, p-valor = {p:.10f}")
```

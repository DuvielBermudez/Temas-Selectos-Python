```python
# ejemplo de codigo para un objeto
class Estudiante:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        return f"Soy {self.nombre} y tengo {self.edad} años"

# Crear un objeto de la clase Estudiante
estudiante1 = Estudiante("María", 21)
print(estudiante1.presentarse())  # Salida: Soy María y tengo 21 años
```

```python
#ejemplo de codigo para un arrays
# Lista de números
calificaciones = [85, 92, 78, 95, 88]

# Acceso por índice
print(calificaciones[0])  # Salida: 85

# Modificación
calificaciones[2] = 80
print(calificaciones)  # Salida: [85, 92, 80, 95, 88]

# Lista heterogénea
datos_mixtos = ["Python", 3.11, True, [1, 2, 3]]
```

```python
#Ejemplo de una serie
import pandas as pd

# Crear una Serie
temperaturas = pd.Series([23, 25, 22, 24, 26],
                         index=['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'])

print(temperaturas['Miércoles'])  # Salida: 22
print(temperaturas.mean())  # Salida: 24.0 (promedio)
```

```python
#ejemplo de diccionario
# Diccionario de información de un estudiante
estudiante = {
    "nombre": "Carlos",
    "edad": 22,
    "carrera": "Ciencias de la Computación",
    "calificaciones": [90, 85, 92]
}

# Acceso por clave
print(estudiante["nombre"])  # Salida: Carlos

# Agregar nuevo par clave-valor
estudiante["semestre"] = 6

# Iteración
for clave, valor in estudiante.items():
    print(f"{clave}: {valor}")
```

```python
#Declarar las variables
n = 2.0 #mol
R = 0.081 #atm*L/molK
T= 298 #K
V = 10 #L

#Plantear la ecuación
P=((n*R*T)/V)
print(f'El valor de la presión para el sistema es: {P:2f} atm')
```

```python
import math
#Declarar las variables
A = 50000000 #s-1
Ea = 75000 #J/molK
T= 350 #K
R = 8.314 #J/molK

#Plantear la ecuación
exponente=(-Ea/(R*T))
k=A*(math.exp(exponente))
print(f'El valor de la constante de velocidad usando la ecuación de Arrhenius es: {k} s-1')
```

Ejercicio 5

```python
import math
#Para leer los datos introducidos por el usuario usamos la función input(),
#pero esta funcion siempre devuelve texto por lo cúal no se puede realizar
#operaciones matemáticas con esta función para resolver este problema usamos
#la función float() la que nos permite convertir el texto a número decimal
rho= float(input('Densidad del flujo Kg/m3 '))
v= float(input('velocidad promedio del fluido m/s '))
D= float(input('Diamétro interno de la tubería m '))
mu= float(input('Viscosidad dinámica Pa/s '))
#Luego planteo la ecuación de Reynolds a resolver con los valores que le
#proporcione al programa
Re=((rho*v*D)/mu)
print(f'el Reynolds es {Re:.2f}')
if Re >= 2300 and Re <= 4000:
  print('Flujo transitorio')
elif Re > 4000:
  print('Flujo turbulento')
else:# <2300 NA
  print('Flujo laminar')
```

```python
#Definiendo las variables del punto crítico del agua
Tc = 647.1 #temperatura crítica en °K
Pc = 22.06 #Presión crítica en MPa

#Solicitar datos de presión y temperatura de su sitema al usuario
T = float(input('Temperatura del sistema:T  [°K] '))
P = float(input('Presión del sistema:P  [MPa] '))
# Determinando el estado del agua
if T<Tc and P<Pc:
  print('El agua se encuentra en fase líquida o vapor')
elif T>Tc and P>Pc:
  print('El agua está en estado super crítico')
elif T<Tc and P>Pc:
  print('El agua se encuentra en condiciones de líquido comprimido')
else:
  print('Las condiciones son cercanas al punto crítico')
```

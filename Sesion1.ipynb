```python
def a_K(T, unidad):
  if unidad == 'C':
    res= T+273.15
    return res
  elif unidad == 'F':
    res=(T-32)*5/9+273.15
    return res
  elif unidad == 'R':
    res =T+5/9
    return res
  elif unidad =='K':
    res=T
    return res
  else :
    print('Dato no valido, verfifique las unidades')
```

```python
a_K(75,'C')
```

Para la presion

```python
def a_P(P, unidad):
  if unidad == 'bar':
    res= P*0.1
    return res
  elif unidad == 'mmHg':
    res=P*0.000133322
    return res
  elif unidad == 'psi':
    res =P*0.00689476
    return res
  elif unidad =='Mpa':
    res=P
    return res
  else :
    print('Dato no valido, verfifique las unidades')
```

```python
a_P(1,'mmHg')
```

Para determinar el estrado del agua

```python
def clasificar_estado (T_K, P_Mpa):
  Tc = 647.1
  Pc=22.06


  if T_K < Tc and P_Mpa < Pc:
     return 'liquido o vapor'
  elif T_K>Tc and P_Mpa>Pc:
     return 'Supercritico'
  elif T_K<Tc and P_Mpa>Pc:
     return 'liquido comprimido'
  else:
     return 'cerca del punto critico'
```

```python
clasificar_estado (345, 21)
```

```python
T=float(input('Temperatura: '))
uT= input('unidad(°C,F,R o K): ')

P=float(input('Presion: '))
uP= input('unidad(bar, mmHg, Psi, atm o Mpa): ')

T_K= a_K(T, uT)
P_Mpa= a_P(P,uP)

print(clasificar_estado(T_K, P_Mpa))
```

```python
capitales = ['#Ciudad Victo*ria', 'cuerna#vaca', 'Morelia!', 'xalapa', 'Tepic', 'hermos*illo',
 'toluca?', 'San? Luis Potosí', '*Mexicali', 'Villahermosa', 'Tuxtla Gutiérrez#',
 'Colima', 'la Paz#', 'Guadalajara', 'C*ampeche!', 'Durango', 'Zacatecas',
 '!Chetumal', 'Ciudad# de México', 'chilpancingo*', 'Chihua#hua', '?Aguascalientes',
 'Oaxaca de Juárez*', '?Puebla', 'Quer?étaro', 'Mérida', 'guanajuato',
 'pachu*ca', '#Saltillo#', 'Culiacán!', 'Tlaxcala', 'monterrey']
```

```python
import re
def limpiar_cadenas(strings):
   result=[]
   for value in strings:
    value = value.strip() #con este cvodigo quita los blancos u otro caracter especificado
    value=re.sub('[#*?!]' , '', value)  #remĺaza todas las ocurrencias de un patron
    value = value.title() #convierte la prmera letra de cada palabra en una cadena
    result.append(value)
   return result
```

```python
capitales_limpias = limpiar_cadenas(capitales)
capitales_limpias
```

```python
capitales_limpias.sort()  #para ordenar alfabeticamente
capitales_limpias
```

```python
ruta_poblacion='https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/poblacion.csv'
```

```python
poblacion =pd.read_csv(ruta_poblacion)
```

```python
poblacion.head()
```

```python
poblacion['1990'] = poblacion['H_1990'] + poblacion['M_1990']
poblacion['2000'] = poblacion['H_2000'] + poblacion['M_2000']
poblacion['2010'] = poblacion['H_2010'] + poblacion['M_2010']
poblacion['2020'] = poblacion['H_2020'] + poblacion['M_2020']
```

```python
poblacion=poblacion.drop(['H_1990','M_1990','H_2000',
                          'M_2000','H_2010','M_2010',
                          'H_2020','M_2020','Regiones'], axis=1)
poblacion
```

```python
poblacion.set_index('Estado')
```

```python


F=['1990','2000','2010','2020']
C=[poblacion['1990'].sum(),poblacion['2000'].sum(),
   poblacion['2010'].sum(),poblacion['2020'].sum()]




pob = pd.DataFrame({'Poblacion' : C}, index=F)


```

```python
pob
```

```python
import matplotlib.pyplot as plt
plt.bar(pob.index, pob['Poblacion'], color = 'blue', alpha=0.6)
plt.title('Poblacion por año ')
plt.ylabel('Poblacion (100,000,000)')
plt.xlabel('Año')
plt.show()
```

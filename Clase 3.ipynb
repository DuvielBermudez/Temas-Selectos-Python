Se usa cuando no sabemos cuantas veces se va a repetir el ciclo solo queremos que se cumpla una condicion 'whiele'

```python
x = 0
while x<5:
  print('x vale:',x)
  x+=1
```

Si por error se va un ciclo al infinito se puede usar 'break' para parar el ciclo y 'continue' permite saltar la linea de codigo

```python
for i in range(1,10):
  if i == 5: #DETIENE EL CICLO AL LLEGARE A 5
    break
  print (i)
```

```python
L=[1,2,None,4,None,5]
suma =0
for i in L:
  if i is None: continue
  suma+=i
suma
```
